```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.preprocessing import OneHotEncoder
```

```python
# Accessing the data
delay_data = pd.read_csv("./data/full_data_flightdelay.csv")
delay = pd.DataFrame(delay_data)
```

```python
# Listing the types of data
delay.dtypes
```

```python
# Looking at what the data looks like
delay.head()
```

```python
# Accesing the categorical features and to perform one hot encoding on.
categorical_features = ["DEP_TIME_BLK", "CARRIER_NAME", "DEPARTING_AIRPORT"]
```

```python
# Creating an encoder
enc = OneHotEncoder(sparse_output = False)
```

```python
#Iterating through the categorical features
for i in range(len(categorical_features)):
    encode_df = pd.DataFrame(enc.fit_transform(delay[categorical_features[i]].values.reshape(-1,1))) 
    encode_df.columns = enc.get_feature_names_out([categorical_features[i]])
    delay = delay.merge(encode_df,left_index = True, right_index = True).drop([categorical_features[i]], axis = 1)
```

```python
# Converting the Previous_Airport category to a binary
delay["PREVIOUS_AIRPORT"] = np.where(delay["PREVIOUS_AIRPORT"] == "NONE", 0, 1)
```

```python
# printing the feature and there respective data type
for i in delay.dtypes:
    for j in delay.columns:
        print(i, j)
```

```python
# Saving the cleaned data
delay.to_csv("./data/clean_data.csv", index = False)
```
